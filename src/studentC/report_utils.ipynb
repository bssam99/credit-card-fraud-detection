```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import (RocCurveDisplay, PrecisionRecallDisplay, confusion_matrix, ConfusionMatrixDisplay)
import os

RESULTS_DIR = "D:\\Queens\\DL-CISC 867-Deep Learning\\Project\\credit-card-fraud-detection\\results"
os.makedirs(RESULTS_DIR, exist_ok=True)

def plot_roc_pr(y_true, y_probs, model_name, results_dir=RESULTS_DIR):
    """Plot and save ROC and PR curves."""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    RocCurveDisplay.from_predictions(y_true, y_probs, ax=ax1, name=model_name)
    ax1.set_title(f"ROC Curve — {model_name}")
    PrecisionRecallDisplay.from_predictions(y_true, y_probs, ax=ax2, name=model_name)
    ax2.set_title(f"PR Curve — {model_name}")
    plt.tight_layout()
    fname = f"{results_dir}/{model_name.replace(' ', '_')}_roc_pr.png"
    plt.savefig(fname, dpi=150)
    plt.close()
    print(f"Saved: {fname}")

def plot_confusion_matrix(y_true, y_pred, model_name, results_dir=RESULTS_DIR):
    """Plot and save confusion matrix."""
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(cm, display_labels=['Legit', 'Fraud'])
    fig, ax = plt.subplots(figsize=(6, 5))
    disp.plot(ax=ax, colorbar=False, cmap='Blues')
    ax.set_title(f"Confusion Matrix — {model_name}")
    plt.tight_layout()
    fname = f"{results_dir}/{model_name.replace(' ', '_')}_confusion_matrix.png"
    plt.savefig(fname, dpi=150)
    plt.close()
    print(f"Saved: {fname}")

def plot_model_comparison(csv_path=f"{RESULTS_DIR}/full_model_comparison.csv"):
    """Bar chart comparing all models."""
    df = pd.read_csv(csv_path)
    fig, ax = plt.subplots(figsize=(10, 5))
    x = np.arange(len(df))
    w = 0.35
    ax.bar(x - w/2, df['roc_auc'], w, label='ROC-AUC', color='steelblue')
    ax.bar(x + w/2, df['pr_auc'],  w, label='PR-AUC',  color='crimson')
    ax.set_xticks(x)
    ax.set_xticklabels(df['name'], rotation=15)
    ax.set_ylim(0, 1.05)
    ax.set_ylabel("Score")
    ax.legend()
    ax.set_title("Model Comparison: ROC-AUC vs PR-AUC")
    plt.tight_layout()
    plt.savefig(f"{RESULTS_DIR}/full_model_comparison.png", dpi=150)
    plt.close()
    print("Model comparison chart saved.")

if __name__ == "__main__":
    print("Report utilities ready.")
```
